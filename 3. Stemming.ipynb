```python
from nltk.stem import PorterStemmer
porter = PorterStemmer()
```

```python
words = ["connects", "connecting", "connect", "connected", "connectings"]
for word in words:
  print(porter.stem(word))
```

```python
words = ["trouble", "Troubling", "troubles", "troubled"]
for word in words:
  print(porter.stem(word))
```

```python
words = ["generalize", "generalization", "general"]
for word in words:
  print(porter.stem(word))
```

```python
from nltk.stem import SnowballStemmer
snowball = SnowballStemmer(language = "english")
```

```python
words = ["generate", "generous", "generation", "generously"]
for word in words:
  print(snowball.stem(word))
```

```python
words = ["generate", "generous", "generation", "generously"]
for word in words:
  print(porter.stem(word))
```

```python
from nltk.stem import RegexpStemmer
regexp = RegexpStemmer("ing$|ed$|es$|able$|tion$")
```

```python
words = ["advisable", "generation", "troubles", "trouble"]
for word in words:
  print(regexp.stem(word))
```
