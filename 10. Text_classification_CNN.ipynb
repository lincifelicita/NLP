```python
def lower_case_conversion(text):
  lower_text = text.lower()
  return lower_text

def remove_urls(text):
    pattern = r'https?://\S+|www\.\S+'
    without_url = re.sub(pattern = pattern, repl = '', string = text)
    return without_url
    
def remove_special(text):
  return re.sub(r'[^a-zA-Z0-9\s]+', '', text)

def remove_numbers(text):
  num_pattern = r'\d+'
  without_number = re.sub(pattern = num_pattern, repl = "", string = text)
  return without_number

#def alpha(text):
#  return re.sub(r'[^a-zA-Z]', ' ', text)
```

```python
df['text'] = df['text'].apply(lower_case_conversion)
df['text'] = df['text'].apply(remove_urls)
df['text'] = df['text'].apply(remove_special)
df['text'] = df['text'].apply(remove_numbers)
df.head()
```
