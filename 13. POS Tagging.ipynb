```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
stop_words = set(stopwords.words('english'))
```

```python
text = 'I love dogs, my pet name is Scooby'
```

```python
tokenized = sent_tokenize(text)
for i in tokenized:
  word_list = word_tokenize(i)
  word_list = [w for w in word_list if not w in stop_words]
  tagged = nltk.pos_tag(word_list)
  print(tagged)
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
doc = nlp(text)
for token in doc:
  print(token, token.pos_)
print('Verbs - ', [token.text for token in doc if token.pos_ == 'VERB'])
```

```python
sentence = 'Microsoft is planning to buy the India startup for $1 billion'
doc = nlp(sentence)
for ent in doc.ents:
  print(ent.text, ent.start_char, ent.end_char, ent.label_)
```

```python
from spacy import displacy
raw_text = '''The Indian Space Research Organisation or is the national space agency in India, headquartered in Bengaluru. 
            It operates under Department of Space which is directly overseen by the Prime Minister of India while 
            Chairman of acts as the executive of as well'''
text1 = nlp(raw_text)
for word in text1.ents:
  print(word.text, word.label_)
```

```python
displacy.render(text1, style = 'ent', jupyter = True)
```
