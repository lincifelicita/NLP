```python
import pandas as pd
from autoviml.Auto_ViML import Auto_ViML
```

```python
train = pd.read_csv('/content/drive/MyDrive/NLP/Train_fake_news.csv')
test = pd.read_csv('/content/drive/MyDrive/NLP/Test_fake_news.csv')
sample = pd.read_csv('/content/drive/MyDrive/NLP/sample submission_fake_news.csv')
```

```python
train.head()
```

```python
train['Labels'].value_counts()
```

```python
train['Text'] = train['Text'] + ' ' + train['Text_Tag'].fillna("").apply(lambda x: ' '.join(x.split(',')))
test['Text'] = test['Text'] + ' ' + test['Text_Tag'].fillna("").apply(lambda x: ' '.join(x.split(',')))
```

```python
train.head()
```

```python
target = 'Labels'
```

```python
m, feats, trainm, testm = Auto_ViML(train, target, test, sample,
                                    scoring_parameter='neg_log_loss',
                                    hyper_param='RS',feature_reduction=True,
                                    KMeans_Featurizer = False,
                                    Boosting_Flag='CatBoost',
                                    Binning_Flag=False,
                                    Add_Poly=0, Stacking_Flag=True,                                    
                                    Imbalanced_Flag=False, 
                                    verbose=3)    
```
