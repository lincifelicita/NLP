##NLTK

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
text = "I love to travel around the world and to play cricket with my friends"
```

```python
text_tokens = word_tokenize(text)
tokens_stop = [word for word in text_tokens if not word in stopwords.words()]
tokens_stop
```

```python
all_stopwords = stopwords.words('english')
all_stopwords.append('love')
text_tokens = word_tokenize(text)
tokens_stop = [word for word in text_tokens if not word in all_stopwords]
tokens_stop
```

```python
text = "I love to travel around the world"
```

```python
x = ['around', 'love']
all_stopwords.extend(x)
text_tokens = word_tokenize(text)
tokens_stop = [word for word in text_tokens if not word in all_stopwords]
tokens_stop
```

```python
all_stopwords = stopwords.words('english')
all_stopwords.remove('to')
text_tokens = word_tokenize(text)
tokens_stop = [word for word in text_tokens if not word in all_stopwords]
tokens_stop
```

## Gensim

```python
import gensim
from gensim.parsing.preprocessing import STOPWORDS
from gensim.parsing.preprocessing import remove_stopwords
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
len(STOPWORDS)
```

```python
text = "Nick likes to play football, however he is not too font of tennis"
```

```python
filtered_sent = remove_stopwords(text)
filtered_sent
```

```python
all_stopwords_gensim = STOPWORDS.union(set(['likes', 'play']))
text_tokens = word_tokenize(text)
tokens_without_sw = [word for word in text_tokens if not word in all_stopwords_gensim]
tokens_without_sw
```

```python
all_stopwords_gensim = STOPWORDS
sw_list = {"not"}
all_stopwords_gensim = STOPWORDS.difference(sw_list)
text_tokens = word_tokenize(text)
tokens_without_sw = [word for word in text_tokens if not word in all_stopwords_gensim]
tokens_without_sw
```
