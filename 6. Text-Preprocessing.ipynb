## Lowercase conversion

## HTML Removal


```python
import re
def remove_html_tags(text):
  html_pattern = r'<.*?>' #greedy match - stops when sees the first match
  without_html = re.sub(pattern = html_pattern, repl = '', string = text)
  return without_html
ex_html_tags = """ <body>
<div>
<h1> Hello! this is an example! </h1>
</div>
</body>
"""
remove_html_tags(ex_html_tags)
```

```python
from bs4 import BeautifulSoup
def remove_html_tags_bs(text):
  parser = BeautifulSoup(text, "html.parser")
  without_html = parser.get_text(Separator = " ")
  return without_html
ex_html_tags = """ <body>
<div>
<h1> Hello! this is an example! </h1>
</div>
</body>
"""
remove_html_tags(ex_html_tags)
```

## URL Removal

```python
def remove_urls(text):
    pattern = r'https?://\S+|www\.\S+'
    without_url = re.sub(pattern = pattern, repl = '', string = text)
    return without_url
ex_url=""" patapatapatapata https://www.google.com"""
remove_urls(ex_url)
```

## Numbers Removal

```python
def remove_numbers(text):
  num_pattern = r'\d+'
  without_number = re.sub(pattern = num_pattern, repl = "", string = text)
  return without_number
ex_num = """ this is a sentence fro removing num666
"""
remove_numbers(ex_num)
```

## Number to word conversion

```python
!pip install num2words
from num2words import num2words
```

```python
def number_to_words(text):
  after_split = text.split()
  for idx in range(len(after_split)):
    if after_split[idx].isdigit():
      after_split[idx] = num2words(after_split[idx])
    num_to_words = ' '.join(after_split)
  return num_to_words
ex_num = """ this is a sentence 34 for 4 removing num 666
"""
number_to_words(ex_num)
```

## Spelling correction using spellchecker

```python
!pip install pyspellchecker
from spellchecker import SpellChecker
spell_corr = SpellChecker()
```

```python
def spell_check(text):
  correct_words = []
  mis_words = spell_corr.unknown(text.split())
  for each_word in text.split():
    if each_word in mis_words:
      right_word = spell_corr.correction(each_word)
      correct_words.append(right_word)
    else:
      correct_words.append(each_word)
  correct_spelling = ' '.join(correct_words)
  return correct_spelling
ex_spell_check = "This is an exmple of corection"
spell_check(ex_spell_check)
```

## Spelling correction using autocorrect

```python
!pip install autocorrect
from autocorrect import Speller
import nltk
nltk.download('punkt')
from nltk import word_tokenize
```

```python
def spell_autocorrect(text):
  correct_spell_words = []
  spell_corr = Speller(lang = 'en')
  for word in word_tokenize(text):
    correct_word = spell_corr(word)
    correct_spell_words.append(correct_word)
  correct_spelling = ' '.join(correct_spell_words)
  return correct_spelling
ex_spell_check = "This is an exmple of corection"
spell_autocorrect(ex_spell_check)
```

```python
def remove_emoticons(text):
  """
  Return :- string after removing emoticons
  Input :- string
  Output :- string
  """
  emoticon_pattern = re.compile(u'(' + u'|'.join(k for k in EMOTICONS) + u')')
  without_emoticons = emoticon_pattern.sub(r'',text)
  return without_emoticons
# example sentence for removing emoticons
ex_emoticons = """
Hello this is a sentence with these 3 emoticons :-) , :-3 , :X
"""
emoticons_result = remove_emoticons(ex_emoticons)
print(f"After removing emoticons :- \n{emoticons_result}")
```

```python
def single_char(text):
  pattern = r'\s[a-zA-Z]\s'
  without_single_char = re.sub(pattern = pattern, repl = '', string = text)
  return without_single_char
ex_single_char=""" She is b  good at painting"""
single_char(ex_single_char)
```

```python
#lower case conversion
df['review'] = df['review'].apply(lower_case_conversion)
df.head()
```

```python
#html tags removal
df['review'] = df['review'].apply(remove_html_tags)
df.head()
```

```python
#removing contractions
contraction_mapping = CONTRACTION_MAP
contractions_pattern = re.compile('({})'.format('|'.join(contraction_mapping.keys())),
flags=re.IGNORECASE|re.DOTALL)
for i in range(len(df)):
  expanded_text = contractions_pattern.sub(expand_contractions, df['review'][i])
  expanded_text = re.sub("'", "", expanded_text)
  df['review'][i] = expanded_text
df.head()
```

```python
#removing single character
df['review'] = df['review'].apply(single_char)
df.head()
```

```python
#removing special characters
df['review'] = df['review'].apply(remove_special)
df.head()
```

```python
#removing stopwords
import gensim
from gensim.parsing.preprocessing import remove_stopwords

for i in range(len(df)):
  df['review'][i] = remove_stopwords(df['review'][i])
df.head()
```

```python
#number to word conversion
df['review'] = df['review'].apply(number_to_words)
df.head()
```

```python
#lemmatization
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk import pos_tag

wordnet = WordNetLemmatizer()
for i in range(len(df)):
  lemm_word = []
  for token,tag in pos_tag(word_tokenize(df['review'][i])):
    pos = tag[0].lower()
    if pos not in ['a', 'r', 'n', 'v']:
      pos = 'n'
    lemm_word.append(wordnet.lemmatize(token, pos))
    df['review'][i] = ' '.join(lemm_word)
df.head()
```
