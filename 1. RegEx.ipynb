```python
import re
```

```python
def fun(text):
  res = re.sub(r'\d+', '', text)
  return res
fun("There are 3 balls in this bag, and 12 in the other one.")
```

```python
def remove_urls(text):
  url_pattern = re.compile(r'https?://\S+|www\.\S+')

  return url_pattern.sub(r'', text)
```

```python
text = "Please refer to link http://lnkd.in/ecnt5yC for the paper"
remove_urls(text)
```

## findall

```python
target_string = "Emma is a basketball player who was born on June 17, 1993. She played 112 matches with scoring average 26.12 points per game. Her weight is 51 kg."
```

```python
re.findall(r"\d+", target_string) # returns a number with one or more didgit and all occurences
```

```python
re.search(r"\d+", target_string) # returns a number with one or more didgit and first occurence
```

```python
res = re.finditer(r"\d+", target_string) # returns iterator obj
```

```python
for i in res:
  print(i.group())
```

```python
target_string = "Jessa is a Python developer. She also gives Python programming training"
```

```python
re.findall(r"\b[Pp]\w+", target_string) # starting with 'p'
```

```python
re.findall(r"\b[Pp]\w+[g]", target_string) # starts with 'p' and ends with 'g'
```

```python
re.findall(r"\b[PpTt]\w+[g]", target_string) # starts with 'p' or 't' and ends with 'g'
```

```python
target_string = "banana, apple, orange"
```

```python
re.findall(r"\b[Aa]\w+|\w+[Aa]\b", target_string) # starts or ends with 'a'
```

```python
x = "1234567890, (123)-567-1234"
re.findall(r"\d{10}", x)
```

```python
re.findall(r"\(\d{3}\)-\d{3}-\d{4}", x)
```

```python
y = "+91 9894962149, 919894962149, 9894962149"
re.findall(r"\b\+91\s\d{10}|\b91\d{10}|\b\d{10}", y)
```

```python
z = "python Python"
re.findall(r"[p]\w+", z, re.I) #ignore case
```

```python
a = 'ML\nand Python'
re.findall(r".+", a, re.S|re.I) #print new line char also
```

## groups

```python
string = 'John has 6 cats but I think my friend Susan has 3 dogs and Mike has 8 fishes'
```

```python
re.findall(r"\b[A-Z]\w+\s\w+\s\d\s\w+", string)
```

```python
re.findall(r"\b[A-Za-z+]+\s\w+\s\d+\s(\w+)", string) # '()' extract portions inside paranthesis
```

```python
info = re.findall(r"\b([A-Za-z]+)+\s(\w+)\s(\d+)\s(\w+)", string) # '()' extract portions inside paranthesis
```

```python
list(zip(*info)) #groups similar words together
```

```python
info = re.search(r"\b([A-Za-z]+)+\s(\w+)\s(\d+)\s(\w+)", string)
```

```python
info.group(0)
```

```python
info.group(1)
```

```python
info.group(1, 3)
```

```python
info.span()
```

```python
info.span(1)
```

```python
re.findall(r"\b(([A-Za-z]+)+\s(\w+)\s(\d+)\s(\w+))", string)
```

## split

```python
s = "My name is maximums and my luck numbers are 12 45 78"
```

```python
re.split(r"\s+", s) #splits based on the pattern given
```

```python
x = "12-34-45-00"
re.split(r"\D", x, maxsplit = 1) #splits the first occurrence of the pattern and the rest are considered as next split
```

```python
x = "12-34-45-00"
re.split(r"\D", x, maxsplit = 2)
```

```python
x = "12,34,45-00"
re.split(r",|-", x) # splits using multiple char
```

```python
t = "PYnative dot.com; is for, Python-developer"
re.split(r"[-,.;\s]\s*", t)
```

```python
re.split(r"[\b\W\b]+", t)
```

```python
text = "12, and45,78and85-17and89-97"
re.split(r"[\b\D\b]+", text)
```

```python
re.split(r"\D+", "12-22-22-.34")
```

```python
re.split(r"(\D+)", "12-22-22-.34") #includes delimiters
```

## substitute

```python
s = "Jessa knows testing and machine learning"
```

```python
a = "   Jessa knows testing    and machine learning"
re.sub(r"\s", r"", a)
```

```python
re.sub(r"^\s+", r"", a)
```

```python
re.sub(r"\s", r"_", s, count = 2) #replaces only for the given count
```
